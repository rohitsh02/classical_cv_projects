```python
# Let's create a function to view image
def imshow(title="title", image=None):
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
# let's improve our imshow function
def imshow(title="Image", image=None, size=10):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size * aspect_ratio, size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

## Transformations in image

```python
# Load image
image = cv2.imread("/content/thisisengineering-raeng-64YrPKiguAE-unsplash.jpg")
```

```python
imshow(image=image)
```

```python
# store height and width of the image
height, width = image.shape[:2]
```

```python
# shift it by quater of the height and width
quarter_height, quarter_width = height/4, width/4
```

```python
# T is our translation matrix
T = np.float32([[1, 0, quarter_width], [0, 1, quarter_height]])
```

```python
# warpAffine to transform the image using the matrix T
img_trans = cv2.warpAffine(image, T, (width, height))
imshow("translate", img_trans)
```

## Rotations

`cv2.getRotationMatrix2D(rotation_centre_x, rotation_centre_y, angle of rotation, scale)`

```python
height, width = image.shape[:2]
# Divide by two to rotate the image around its centre
rotational_matrix = cv2.getRotationMatrix2D((width/2, height/2), 90, 1)
# input our image the rotation matrix and our desired final width and height
rotated_image = cv2.warpAffine(image, rotational_matrix, (width, height))
imshow(image=rotated_image)
```

```python
# Rotations with cv2.transpose(less flexible)
rotated_image = cv2.transpose(image)
imshow(image = rotated_image)
```

```python
# Let's now flip horizontally
flipped = cv2.flip(image, 2)
imshow(image=flipped)
```
